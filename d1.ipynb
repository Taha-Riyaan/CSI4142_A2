*   Weekly AI Upskilling hrs: Hours per week spent learning new AI skills
*   Productivity Score: Self-reported productivity score (Higher = more productive)
*   Burnout Score: Composite burnout indicator (Higher = more burned out) 
*   Job Satisfaction 1-5: Job satisfaction rating on 1-5 scale (1 = very dissatisfied, 5 = very satisfied)
*   Fear Of AI Replacement: Perceived threat of AI replacing their job (Low / Medium / High)
*   Attrition Risk: Predicted likelihood of employees leaving their job within 12 months (Low / Medium / High)

### Presence Errors

Presence check:
Checks that data entered in mandatory fields are not blank.

--------------------------------------------------------------------------------

```python
#cell 2 - How you introduced such error
# Introduced error by setting fear of AI replacement to "massive" for 75 random employees
randomEmployees = np.random.choice(1500, size=75, replace=False)
np.random.seed(37) # for reproducibility
for i in range(75):
    df.loc[randomEmployees[i], "job_role"] = None
# print("Introduced errors at indices:", randomEmployees) #for testing purposes, to know where the errors are
```

--------------------------------------------------------------------------------

```python
#cell 3 - Checker code
# identify rows that have any NaN/None or empty string in any column
nullMask = df.isnull().any(axis=1)
emptyMask = (df == '').any(axis=1)

# rows where at least one column is null/empty
errorRows = df[nullMask | emptyMask]
errorIndices = errorRows.index.tolist()
# print("Indices of rows with errors:", errorIndices)
# print("problematic number of rows:", len(errorIndices))
# errorIndices, errorRows.head()
```

--------------------------------------------------------------------------------

```python
#cell 4 - report of findings
errorsFoundExample = errorIndices[:3] # get first 3 errors found for demo purposes
print("Original values:")
print(ogdf["job_role"].loc[errorsFoundExample])
print("\n")
print("Modified values:")
print(df["job_role"].loc[errorsFoundExample])
```

### Look-up Errors

Look-Up check:
Checks that data entered that has a limited set of values has a proper value using a table.

--------------------------------------------------------------------------------

```python
#cell 2 - How you introduced such error
# Introduced error by setting fear of AI replacement to "massive" for 75 random employees
randomEmployees = np.random.choice(1500, size=75, replace=False)
np.random.seed(42) # for reproducibility
for i in range(75):
    df.loc[randomEmployees[i], "fear_of_ai_replacement"] = "massive"
# print("Introduced errors at indices:", randomEmployees) #for testing purposes, to know where the errors are
```

--------------------------------------------------------------------------------

```python
#cell 3 - Checker code
# Check for invalid values in columns that specified acceptable values
errorRows = [] # to store rows with errors for reporting
adoptValues = ["Experimenting","Integrating", "Optimizing", "AI-First"] # values for ai_adoption_stage
adoptInvalid = df.loc[~df["ai_adoption_stage"].isin(adoptValues)].index.tolist() # get indices of rows where ai_adoption_stage is not in the valid values
# print("Invalid ai_adoption_stage values at indices:", adoptInvalid)
errorRows.extend(adoptInvalid)

replacementValues = ["Low", "Medium", "High"] # values for fear_of_ai_replacement
replacementInvalid = df.loc[~df["fear_of_ai_replacement"].isin(replacementValues)].index.tolist() # get indices of rows where fear_of_ai_replacement is not in the valid values
errorRows.extend(replacementInvalid)
# print("Invalid row indices:", replacementInvalid) # for testing purposes, to check that errors are caught
# print("Number of rows with errors found:", len(errorRows))

attritionValues = ["Low", "Medium", "High"] # values for attrition_risk
attritionInvalid = df.loc[~df["attrition_risk"].isin(attritionValues)].index.tolist() # get indices of rows where attrition_risk is not in the valid values
errorRows.extend(attritionInvalid)
```

--------------------------------------------------------------------------------

```python
#cell 4 - report of findings
errorsFoundExample = replacementInvalid[:3] # get first 3 errors found for demo purposes
print("Original values:")
print(ogdf["fear_of_ai_replacement"].loc[errorsFoundExample])
print("\n")
print("Modified values:")
print(df["fear_of_ai_replacement"].loc[errorsFoundExample])
```
